```python
import pandas as pd
import numpy as np
```

```python
csv= '/home/mei/nas/docker/thesis/data/csv/'
```

```python
print('==> Loading data drug.csv...')
drug = pd.read_csv(csv + 'drug.csv')
drug=drug.rename(columns={'patientunitstayid': 'patient'})
```

```python
n=len(list(drug['patient'].unique()))
r=len(drug)
print("There are {} unique patientunitstayid in the selected patients with the total {} records.".format(n,r))
```

## select top 100 drugs

```python
top100_drugs = drug['drugname'].value_counts().head(100).index
drug= drug[drug['drugname'].isin(top100_drugs)]
print('==> Top 100 drugs selected.')
print("there are {} patients in the top 100 drugs and {} records.".format(len(list(drug['patient'].unique())),len(drug)))

df_flat=pd.read_csv(csv+ 'preprocessed_flat.csv')
common_ids = set(df_flat['patient']).intersection(set(drug['patient']))

flat = df_flat[df_flat['patient'].isin(common_ids)]
drug=drug[drug['patient'].isin(common_ids)]
```

```python
one_hot = pd.get_dummies(drug['drugname'])
drug=drug.join(one_hot)
drug=drug.drop(columns=['drugname'])
drug = drug.astype(int)
drug_=drug.groupby('patient').max()
```

```python
flat_drug = pd.merge(flat, drug_, on='patient', how='left') 
```

```python
flat_drug.to_csv(csv + 'preprocessed_flat_drug.csv', index=False)
drug_.to_csv(csv + 'preprocessed_drug.csv', index=False)
print('==> preprocessed_flat_drug.csv saved.')
print('==> preprocessed_drug.csv saved.')
```
