```python
import numpy as np
import pandas as pd
import os
```

```python
csv= '/home/mei/nas/docker/thesis/data/csv/'
hdf= '/home/mei/nas/docker/thesis/data/hdf/'
```

```python
def convert_csv_to_hdf(csv_files, hdf_path, mode="w"):
    """
    param:
    - csv_files (dict): {key: csv_file_path}
    - hdf_path (str): 
    - mode (str): 'w' (default) or 'a'
    """
    os.makedirs(hdf_path, exist_ok=True)
    for key, csv_file in csv_files.items():
        if not os.path.exists(csv_file):
            print(f"file {csv_file} not found")
            continue
        
        print(f"process {csv_file} and save to HDF5('{key}')")
        
        df = pd.read_csv(csv_file)
        
        hdf_file = os.path.join(hdf_path, f"final_{key}.h5")
        df.to_hdf(hdf_file, key=key, mode=mode, complevel=5, complib="zlib")
    
    print(f"all the files are save to {hdf_path}")
```

```python
csv_files = {
    "flat": os.path.join(csv, "final_flat_drug.csv"),
    "labels": os.path.join(csv, "final_labels.csv"),
    "diagnoses": os.path.join(csv, "final_diagnoses.csv")
}


convert_csv_to_hdf(csv_files, hdf)
```
