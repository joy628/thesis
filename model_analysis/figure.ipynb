```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
categories = [
    "Emergency Department", "Floor", "Operating Room", "ICU to SDU", "Direct Admit",
    "Recovery Room", "Acute Care/Floor", "Step-Down Unit (SDU)", "ICU", "Other Hospital",
    "Other ICU", "PACU", "Chest Pain Center", "Observation", "Other"
]
counts = [
    89594, 24368, 24305, 13827, 12672, 7844, 5604, 5450, 5439, 4323,
    4264, 1714, 1090, 336, 19
]

# 
counts, categories = zip(*sorted(zip(counts, categories), reverse=True))

N = len(counts)
# 
cmap = plt.cm.Blues
color_values = np.linspace(0.60, 0.05, N)  # 
colors = cmap(color_values)

plt.figure(figsize=(12, 7))
bars = plt.bar(categories, counts, color=colors)

# 
for bar, count in zip(bars, counts):
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height(),
        f'{count}',
        ha='center',
        va='bottom',
        fontsize=10
    )

plt.ylabel('Count')
plt.xlabel('Category')
plt.title('Admission Scoures counts')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
categories = [
    'Cardiovascular (R)', 'Endocrine (R)', 'Pulmonary', 'Neurologic', 'Hematology',
    'Renal (R)', 'Gastrointestinal (R)', 'Rheumatic', 'Infectious Disease (R)'
]
counts = [12341, 3258, 2310, 1807, 1803, 1350, 1232, 165, 150]
percents = [50.54, 13.34, 9.46, 7.40, 7.38, 5.53, 5.05, 0.68, 0.61]  # 

plt.figure(figsize=(12, 6))
bars = plt.bar(categories, counts, color=plt.cm.Blues(np.linspace(0.65, 0.16, len(categories))))

for bar, count, pct in zip(bars, counts, percents):
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        f'{count}\n({pct:.2f}%)',
        ha='center', va='bottom', fontsize=10
    )

plt.ylabel('Count')
plt.title('Top-level Disease Categories in Past History')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```
